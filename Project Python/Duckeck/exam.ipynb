Probelm 2: Use numpy to read the file rohr1.dat and data analysis

```python
import numpy as np

data = np.loadtxt("rohr1.dat")
maximum = np.max(data)
minimum = np.min(data)
NumberThatLargerThanMinus25 = np.sum(data > -25)
median = np.median(data)

print("Maximum:", maximum)
print("Minimum:", minimum)
print("Larger than -25:", NumberThatLargerThanMinus25)
print("Median:", median)
```
